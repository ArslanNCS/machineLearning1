Заменим аномальные значения в числовых столбцах на медианые

```python
# заменяем месячный баланс, который меньше нуля или больше милиона на медианый
df['Monthly_Balance'] = df['Monthly_Balance'].apply(
        lambda x: df['Monthly_Balance'].median() if x < 0 or x > 1000000 else x
    )
```

```python
# количество займов не должно быть больше 100 и меньше 0
df['Num_of_Loan'] = df['Num_of_Loan'].apply(
        lambda x: df['Num_of_Loan'].median() if x < 0 or x > 100 else x
    )
```

```python
# заменяем годовой баланс, который меньше нуля или больше милиона на медианый
df['Annual_Income'] = df['Annual_Income'].apply(
        lambda x: df['Annual_Income'].median() if x < 0 or x > 1000000 else x
    )
```

```python
# количество банковских аккаунтов не должно быть больше 15 и меньше 0
df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].apply(
        lambda x: df['Num_Bank_Accounts'].median() if x < 0 or x > 15 else x
    )
```

```python
# количество кредитных карт не должно быть больше 15 и меньше 0
df['Num_Credit_Card'] = df['Num_Credit_Card'].apply(
        lambda x: df['Num_Credit_Card'].median() if x < 0 or x > 15 else x
    )
```

```python
# задержка об оплате не должна быть меньше 0
df['Delay_from_due_date'] = df['Delay_from_due_date'].apply(
        lambda x: df['Delay_from_due_date'].median() if x < 0 else x
    )
```

## Обучение моделей

преобразуем данные типа object к бинарным

```python
X = pd.get_dummies(df, columns=df.select_dtypes(include=['object']).columns)
```

```python
X = X.astype({col: 'int' for col in X.select_dtypes(include=['bool']).columns})
```

```python
y = X['Credit_Score']
```

Разделяем выборку на обучающую и тестовую

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### KNeighborsClassifier (Классификация ближайших соседей)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_pred_test = knn.predict(X_test)
```

```python
y_test_predict = knn.predict(X_test)
```

### Gaussian Naive Bayes (НАИВНЫЙ БАЙЕСОВСКИЙ КЛАССИФИКАТОР)

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_pred_test = gnb.predict(X_test)
```

### Random Forest Classifier (Классификатор дерева решений)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc  = RandomForestClassifier(criterion = 'gini',                                   
                                    max_depth = 5,
                                    random_state = 42,
                                    n_jobs = -1)

rfc.fit(X_train, y_train)
rfc_pred_test = rfc.predict(X_test)
```

### Logistic Regression (Логистическая регрессия)

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
lr = LogisticRegression(random_state = 42)
lr.fit(X_train, y_train)
```

```python
lr_pred_test = lr.predict(X_test)
```

## Метрики классификации

```python
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, precision_score, recall_score, f1_score, confusion_matrix, r2_score
from sklearn.metrics import classification_report
```

```python
classifiers =["KNeighborsClassifier", "Gaussian Naive Bayes", "Random Forest Classifier", "Logistic Regression"]

log_cols = ["Classifier", "Accuracy", "Precision", "Recall", "F1"]
log = pd.DataFrame(columns=log_cols)

log["Classifier"] = classifiers
```

```python
rfc_accuracy = accuracy_score(y_test, rfc_pred_test)
lr_accuracy = accuracy_score(y_test, lr_pred_test)
gnb_accuracy = accuracy_score(y_test, gnb_pred_test)
knn_accuracy = accuracy_score(y_test, y_test_predict, normalize=True)
log["Accuracy"] = [knn_accuracy, gnb_accuracy, rfc_accuracy, lr_accuracy]
```

```python
rfc_precision = precision_score(y_test, rfc_pred_test, average='macro')
lr_precision = precision_score(y_test, lr_pred_test, average='micro')
gnb_precision = precision_score(y_test, gnb_pred_test, average='micro')
knn_precision = precision_score(y_test, y_test_predict, average='micro')
log["Precision"] = [knn_accuracy, gnb_accuracy, rfc_accuracy, lr_accuracy]
```

```python
rfc_recall = recall_score(y_test, rfc_pred_test, average='macro')
lr_recall = recall_score(y_test, lr_pred_test, average='micro')
gnb_recall = recall_score(y_test, gnb_pred_test, average='micro')
knn_recall = recall_score(y_test, y_test_predict, average='micro')
log["Recall"] = [knn_recall, gnb_recall, rfc_recall, lr_recall]
```

```python
# F1 = 2 * (Precision * Recall) / (Precision + Recall)
knn_f1 = 2 * (knn_precision * knn_recall) / (knn_precision + knn_recall)
gnb_f1 = 2 * (gnb_precision * gnb_recall) / (gnb_precision + gnb_recall)
rfc_f1 = 2 * (rfc_precision * rfc_recall) / (rfc_precision + rfc_recall)
lr_f1 = 2 * (lr_precision * lr_recall) / (lr_precision + lr_recall)
log["F1"] = [knn_f1, gnb_f1, rfc_f1, lr_f1]
```
