• NumofDelayed_Payment: Общее количество случаев просроченных платежей, совершенных клиентом.

• ChangedCreditLimit: Изменения, внесенные в кредитный лимит клиента.

• NumCreditInquiries: Количество запросов, сделанных по поводу кредита клиента.

• Credit_Mix: Разнообразие типов кредита, которые использует клиент (например, займы, кредитные карты).

• Outstanding_Debt: Общая сумма долга, которую клиент в настоящее время должен.

• CreditUtilizationRatio: Пропорция использованного кредита по сравнению с общим кредитным лимитом.

• CreditHistoryAge: Продолжительность кредитной истории клиента.

• PaymentofMin_Amount: Указывает, платит ли клиент минимально необходимую сумму каждый месяц.

• TotalEMIper_month: Общая сумма равных ежемесячных платежей (EMI), выплачиваемых клиентом.

• Amountinvestedmonthly: Ежемесячная сумма инвестиций, сделанных клиентом.

• Payment_Behaviour: Привычки и тенденции клиента в отношении платежей.

• Monthly_Balance: Остаток на счете клиента в конце каждого месяца.

• Credit_Score: Кредитный рейтинг клиента (целевое значение: "Хороший", "Плохой", "Стандартный").").		"Poor," "Standard").

## Предобработка данных

```python
# заменяем анамальные значения на медианные
med_Num_Bank_Accounts_train = df_train['Num_Bank_Accounts'].median()
med_Num_Bank_Accounts_test = df_test['Num_Bank_Accounts'].median()

def to_do_better_Num_Bank_Accounts_train(Num_Bank_Accounts):  
    if Num_Bank_Accounts < 0 or Num_Bank_Accounts >= 20:
        return med_Num_Bank_Accounts_train
    else:
        return Num_Bank_Accounts
        
def to_do_better_Num_Bank_Accounts_test(Num_Bank_Accounts):  
    if Num_Bank_Accounts < 0 or Num_Bank_Accounts >= 20:
        return med_Num_Bank_Accounts_test
    else:
        return Num_Bank_Accounts
        
df_train['Num_Bank_Accounts'] = df_train['Num_Bank_Accounts'].apply(to_do_better_Num_Bank_Accounts_train)
df_test['Num_Bank_Accounts'] = df_test['Num_Bank_Accounts'].apply(to_do_better_Num_Bank_Accounts_test)
```

```python
# заменяем кол во кредитных кард

def to_do_better_Num_Credit_Card_train(Num_Credit_Card):  
    if Num_Credit_Card < 0 or Num_Credit_Card > 8:
        return df_train['Num_Credit_Card'].median()
    else:
        return Num_Credit_Card
        
def to_do_better_Num_Credit_Card_test(Num_Credit_Card):  
    if Num_Credit_Card < 0 or Num_Credit_Card > 8:
        return df_test['Num_Credit_Card'].median()
    else:
        return Num_Credit_Card
        
df_train['Num_Credit_Card'] = df_train['Num_Credit_Card'].apply(to_do_better_Num_Credit_Card_train)
df_test['Num_Credit_Card'] = df_test['Num_Credit_Card'].apply(to_do_better_Num_Credit_Card_test)
```

```python
# заменяем процентную ставку

def to_do_better_Interest_Rate_train(Interest_Rate):  
    if Interest_Rate < 0 or Interest_Rate > 40:
        return df_train['Interest_Rate'].median()
    else:
        return Interest_Rate
        
def to_do_better_Interest_Rate_test(Interest_Rate):  
    if Interest_Rate < 0 or Interest_Rate > 40:
        return df_test['Interest_Rate'].median()
    else:
        return Interest_Rate
        
df_train['Interest_Rate'] = df_train['Interest_Rate'].apply(to_do_better_Interest_Rate_train)
df_test['Interest_Rate'] = df_test['Interest_Rate'].apply(to_do_better_Interest_Rate_test)
```

```python
# заменяем кол-во запросов по кредитной карте

def to_do_better_Num_Credit_Inquiries_train(value):  
    if value < 0 or value > 100:
        return df_train['Num_Credit_Inquiries'].median()
    else:
        return value
        
def to_do_better_Num_Credit_Inquiries_test(value):  
    if value < 0 or value > 100:
        return df_test['Num_Credit_Inquiries'].median()
    else:
        return value
        
df_train['Num_Credit_Inquiries'] = df_train['Num_Credit_Inquiries'].apply(to_do_better_Num_Credit_Inquiries_train)
df_test['Num_Credit_Inquiries'] = df_test['Num_Credit_Inquiries'].apply(to_do_better_Num_Credit_Inquiries_test)
```

```python
# заменяем сумму к погашению

def to_do_better_Total_EMI_per_month_train(value):
    if value < 0 or value > 2000:
        return df_train['Total_EMI_per_month'].median()
    else:
        return value
        
def to_do_better_Total_EMI_per_month_test(value): 
    if value < 0 or value > 2000:
        return df_test['Total_EMI_per_month'].median()
    else:
        return value
        
df_train['Total_EMI_per_month'] = df_train['Total_EMI_per_month'].apply(to_do_better_Total_EMI_per_month_train)
df_test['Total_EMI_per_month'] = df_test['Total_EMI_per_month'].apply(to_do_better_Total_EMI_per_month_test)
```
